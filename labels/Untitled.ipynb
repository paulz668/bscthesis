```python
import json
import matplotlib.pyplot as plt
```

```python
name = "labels.json"

with open(name, 'r') as file:
    labels = json.load(file)
```

```python
# Set the font family to 'serif' (the default font used by LaTeX)
plt.rc('font', family='serif')

# Create a figure with a grid of subplots
num_rows = len(labels)
num_columns = max(len(data_dict) for data_dict in labels.values())
fig, axes = plt.subplots(num_rows, num_columns, figsize=(10, 5), sharey=True, sharex=True)

for i, (dict_label, data_dict) in enumerate(labels.items()):
    # Create a subplot for each sub-dictionary in a separate row
    for j, (label, values) in enumerate(data_dict.items()):
        x = [1, 2, 3]
        ax = axes[i, j]
        ax.bar(x, values, label=label)

# Create an "invisible" subplot for labeling the columns
column_labels = ['n = 1', 'n = 2', 'n = 3', 'n = 5', 'n = 10']
for i in range(len(column_labels)):
    ax = axes[0, i]
    ax.set_title(column_labels[i], loc='center')
    
row_labels = ['SLM (t=0.001)', 'SLM (t=0.002)', 'SLM (t=0.003)', 'STBML', 'DTBML']
# Add row titles to the left of each row
for i, row_label in enumerate(row_labels):
    axes[i, 0].set_ylabel(row_label, rotation=45, va='center', labelpad=30)

fig.suptitle("Comparison of label frequency per labelling scheme and forecasting period")
    
# Adjust the spacing between subplots
plt.tight_layout()

plt.savefig("labelvisuals.png")

plt.show()
```
